```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-paper')
import numpy as np
```

## Distribution

The distribution for velocities follows a boltzmann with the kinetic energy as the factor. When normalized, the probability distrubition is this:
$$
g(v)=\sqrt{\frac{2\pi k_B T}{m}}e^{-mv^2 / 2 k_B T}
$$

The expected speed for this distribution is

$$
\langle \left|v\right| \rangle = \sqrt{\frac{2 k_B T}{\pi m}}
$$

```python
def g(m, T, v):
    kB = 1.38064852e-23 # J / K
    return np.sqrt(2 * np.pi * kB * T / m) * np.exp(-m * v**2 / (2*kB*T))

def expected_speed(m, T):
    kB = 1.38064852e-23 # J / K
    return np.sqrt(2 * np.pi * kB * T / m)
```

```python
M = 28.01340 # g/mol
m = M*1e-3 # kg in one mole
T = [100, 300, 3000]
v = np.linspace(-4e-9, 4e-9, 1000)
for t in T:
    plt.plot(v, g(m, t, v), label='{} K'.format(t))
    print('Expected speed for T={}K : {} m/s'.format(t, expected_speed(m, t)))
plt.legend()
plt.xlabel('v (m/s)')
plt.ylabel('P(v)')
```
